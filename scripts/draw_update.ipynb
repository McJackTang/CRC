```python
def simplify_ratio(n, m):
    g = gcd(n, m)
    return (n // g, m // g)  # Return the simplified (n, m)
def adjust_phase_to_0_mpi(phi_rr):
    return np.unwrap(phi_rr)  # Use np.unwrap() to handle phase jumps
def get_hr(y, fs=60, min=30, max=180):
    # fs: sampling frequency
    p, q = welch(y, fs, nfft=int(1e6/fs), nperseg=np.min((len(y)-1, 512)))
    return p[(p>min/60)&(p<max/60)][np.argmax(q[(p>min/60)&(p<max/60)])]*60 
from scipy.fft import fft, fftfreq
def calculate_dynamic_n_m_v2(hr_segment, rr_segment, fs=60):
    """
    Dynamically calculate n:m:
      1) Estimate heart rate HR_bpm (30180 bpm) using get_hr
      2) Estimate respiratory rate VR_bpm (660 bpm) using get_hr
      3) ratio = HR_bpm/VR_bpm, round to nearest integer as n, m=1

    Arguments:
      hr_segment -- a segment of ECG or pulse signal (numpy array)
      rr_segment -- a segment of respiratory or volume signal (numpy array)
      fs         -- sampling rate (Hz)
    Returns:
      (n, m)     -- approximate integer ratio
    """
    # 1) Heart rate: 30180 bpm
    HR_bpm = get_hr(hr_segment, fs, min=30, max=180)
    # 2) Respiratory rate: 660 bpm
    VR_bpm = get_hr(rr_segment, fs, min=6,  max=30)
    
    # 3) Ratio rounded
    ratio = HR_bpm / VR_bpm
    n = int(np.round(ratio))
    m = 1
    
    return ratio
def calculate_gamma(sync,ratio_0):
    # Initialize n_m_sync_dict as an empty dictionary
    n_m_sync_dict = {}
    
    n=ratio_0
    # For m=1
    mm=1
    for nn in range((n-1),(n+1)+1):
        if nn <= 1:
            continue # Skip nn=1
        phi_rr_mod=np.mod(sync,2*np.pi*mm)/(2*np.pi)
        window_psi=(2*np.pi/mm) *(np.mod((phi_rr_mod * nn), mm))
        
        cos_sum = (np.sum((np.cos(window_psi)))) / len(window_psi)
        sin_sum = (np.sum((np.sin(window_psi)))) / len(window_psi)
        # Calculate synchronization degree
        gamma = cos_sum**2 + sin_sum**2
        n_m_sync_dict[(nn, mm)] = gamma 
    
    # For m=2
    mm=2
    for nn in range((n-1)*2+1,(n+1)*2):
        phi_rr_mod=np.mod(sync,2*np.pi*mm)/(2*np.pi)
        window_psi=(2*np.pi/mm) *(np.mod((phi_rr_mod * nn), mm))
        
        cos_sum = (np.sum((np.cos(window_psi)))) / len(window_psi)
        sin_sum = (np.sum((np.sin(window_psi)))) / len(window_psi)
        # Calculate synchronization degree
        gamma = cos_sum**2 + sin_sum**2
        n_m_sync_dict[(nn, mm)] = gamma 
    
    # For m=3
    mm=3
    for nn in range((n-1)*3+1,(n+1)*3):
        phi_rr_mod=np.mod(sync,2*np.pi*mm)/(2*np.pi)
        window_psi=(2*np.pi/mm) *(np.mod((phi_rr_mod * nn), mm))
        
        cos_sum = (np.sum((np.cos(window_psi)))) / len(window_psi)
        sin_sum = (np.sum((np.sin(window_psi)))) / len(window_psi)
        # Calculate synchronization degree
        gamma = cos_sum**2 + sin_sum**2
        n_m_sync_dict[(nn, mm)] = gamma 
    
    max_key, max_gamma = max(n_m_sync_dict.items(), key=lambda item: item[1])
    
    
    return max_key, max_gamma
def calculate_dynamic_n_m(hr_segment, rr_segment):
    """Dynamically adjust n:m based on peak detection"""
    # Get peaks in hr_segment and rr_segment
    hr_peaks, _ = find_peaks(hr_segment, width=10)
    rr_peaks, _ = find_peaks(rr_segment, height=0.5)

    n = len(hr_peaks)  
    m = len(rr_peaks)  

    return round(n/m)
def bandpass_filter(data, lowcut=0.5, highcut=3, fs=30, order=3):
    """Apply a bandpass filter to the data."""
    
    if fs is None or fs <= 0:
        return np.zeros_like(data)
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)
def synchrogram_1(hr_signal, rr_signal, title=""):
    
    hr_min, hr_max = hr_signal.min(), hr_signal.max()
    rr_min, rr_max = rr_signal.min(), rr_signal.max()

    hr_norm = 2*(hr_signal - hr_min) / (hr_max - hr_min)-1
    rr_norm = 2*(rr_signal - rr_min) / (rr_max - rr_min)-1
    # Calculate phase of slower signal (rr_signal) using Hilbert transform
    phi_rr = np.angle(hilbert(rr_norm))
    phi_rr = adjust_phase_to_0_mpi(phi_rr)  # Adjust phase to [0, 2m] range
    phi_rr_1 = np.mod(phi_rr, 2 * np.pi )  # Limit phase to [0, 2]
    phi_rr_2 = np.mod(phi_rr, 4 * np.pi)  # Limit phase to [0, 4]
    phi_rr_3 = np.mod(phi_rr, 6 * np.pi )  # Limit phase to [0, 6]
    # Find peak positions of faster signal (hr_signal)
    peaks, _ = find_peaks(hr_norm,width=8)  # Find peaks in hr_signal
    print(len(peaks))
    #peaks_rr = find_peaks(rr_norm)[0]  # Find peaks in rr_signal

    # Extract phase of rr_signal at peak positions
    sync= phi_rr[peaks]  # Extract phase of rr_signal
    sync_1 = phi_rr_1[peaks]
    sync_2 = phi_rr_2[peaks]  # Extract phase of rr_signal
    sync_3 = phi_rr_3[peaks]  # Extract phase of rr_signal
    
    window_size = 600
    ratio_data = []  # To save (peak position, (n, m))
    for i in range(0,len(peaks)): # Calculate initial guess of n:m at each peak
        p=peaks[i]
        start = max(0, p - window_size)
        end = min(len(hr_signal), p + window_size)
        hr_seg_local = hr_signal[start:end]
        rr_seg_local = rr_signal[start:end]
        local_n= calculate_dynamic_n_m(hr_seg_local, rr_seg_local)
        ratio_data.append((i, local_n ))
    print(ratio_data)
    gamma_all = []
    max_n_m=[]
    
    N=50
    
    for k in range(N//2,len(peaks)-N//2):
        # Take relative phase in sliding window
        start = int( k - N//2+1)
        end = int(k + N//2+1)
        window_sync = sync[start:end]
        ratio_cur=ratio_data[k][1]
        n_m,gamma_cur=calculate_gamma(window_sync,ratio_cur)
        gamma_all.append(gamma_cur)
        max_n_m.append(n_m)
        
    
    print(f"len(psi_plus):{len(sync)}")
    max_n_m = [simplify_ratio(n, m) for n, m in max_n_m]
    # === Calculate synchronization index ===
    
    fs = 60  # Actual sampling rate
    
    peaks_trimmed = peaks[N//2:len(peaks)-N//2]  # Peak positions corresponding to gamma_all

    time = np.arange(len(hr_norm)) / fs 
    # Plot results
    fig, axes = plt.subplots(3, 1, figsize=(20, 9))  
    min_val = min(gamma_all)
    max_val = max(gamma_all)
    # Convert peaks to seconds
    time_peaks = peaks / fs
    # Plot synchronization diagram
    # First plot: BVP and RR signals
    axes[0].plot(time, hr_norm, color=colors['red'], label='BVP', linewidth=1.5)
    axes[0].plot(time, rr_norm, color=colors['blue'], label='RESP', linewidth=1.5)
    axes[0].set_title("Synchronization of BVP and Respiration Signals ", fontsize=25)
    axes[0].legend(loc='best', fontsize=16)
    axes[0].set_ylabel("Normalized Amplitude", fontsize=16)
    axes[0].grid(True, alpha=0.3)
    axes[0].tick_params(axis='both', labelsize=20)
    axes[0].set_xlim([time[0], time[-1]])
    
    # Second plot: Synchrogram
    axes[1].plot(time_peaks, sync_1/(2*np.pi), "o", markersize=4, color='black', alpha=0.7)
    #plt.rcParams["text.usetex"] = True
    axes[1].set_title(r'Synchrogram', fontsize=25)
    #plt.rcParams["text.usetex"] = False
    axes[1].set_ylabel("$\mathrm{\psi}_1(t_k)$", fontsize=25, usetex=True, fontweight='bold')
    axes[1].grid(True, alpha=0.3)
    axes[1].tick_params(axis='both', labelsize=20)
    axes[1].set_xlim([time[0], time[-1]])
    
    # Add rectangle to highlight 80-95s region in synchrogram
    axes[1].axvspan(80, 95, alpha=0.2, color='gray')
    
    # Third plot: Synchronization degree
    time_peaks_trimmed = time_peaks[N // 2 : len(peaks) - N // 2]
    scaled_gamma_all = [val * 100 for val in gamma_all]
    
    scaled_min_val = min_val * 100
    scaled_max_val = max_val * 100
    start_tick = math.floor(scaled_min_val * 10) / 10
    end_tick = math.ceil(scaled_max_val * 10) / 10
    
    axes[2].set_ylim([max(0, start_tick-5), min(100, end_tick+5)])
    axes[2].set_yticks(np.arange(int(max(0, start_tick-5)), int(min(100, end_tick+5))+1, 10))
    axes[2].set_xlim([time[0], time[-1]])
    axes[2].plot(time_peaks_trimmed, scaled_gamma_all, '-', color=colors['orange'], linewidth=2)
    
    # Add rectangle to highlight 80-95s region in sync degree plot
    axes[2].axvspan(80, 92, alpha=0.2, color='gray')
    
    axes[2].set_title("Synchronization Degree over Time", fontsize=25)
    axes[2].set_xlabel("Time (s)", fontsize=25)
    axes[2].set_ylabel("Synchronization Degree (%)", fontsize=16)
    axes[2].grid(True, alpha=0.3)
    axes[2].tick_params(axis='both', labelsize=20)

    
    plt.tight_layout()

    plt.savefig("synchrogram_1.pdf", dpi=300, bbox_inches='tight')
    plt.show()

hr_pattern = r"path to 060201_v01_hr_0.npy"
rr_pattern = r"path to 060201_v01_rr_0.npy"
bvp_signal= np.load(hr_pattern)
rr_signal = np.load(rr_pattern)
bvp_signal = bandpass_filter(bvp_signal, lowcut=0.5, highcut=3, fs=60, order=3)
rr_signal = bandpass_filter(rr_signal, lowcut=0.1, highcut=0.5, fs=60, order=3)
synchrogram_1(bvp_signal, rr_signal)
```

# Bland_altman

```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde, pearsonr
from scipy.interpolate import UnivariateSpline
from scipy.signal import welch, butter, filtfilt

# =======================
# 
# =======================
def bandpass_filter(data, lowcut=0.5, highcut=3, fs=30, order=3):
    b, a = butter(order, [lowcut, highcut], fs=fs, btype='band')
    return filtfilt(b, a, data)

def get_hr(y, sr=30, min=30, max=180):
    p, q = welch(y, sr, nfft=1e5/sr, nperseg=np.min((len(y)-1, 256)))
    mask = (p > min/60) & (p < max/60)
    return p[mask][np.argmax(q[mask])] * 60

def extract_group(df, group_keys):
    mask = False
    for key in group_keys:
        mask = mask | df['file'].str.contains(key)
    return df[mask].sort_values('file').reset_index(drop=True)

def calc_metrics(gold, new):
    errors = new - gold
    mae = np.mean(np.abs(errors))
    mae_std = np.std(np.abs(errors))
    mape = np.mean(np.abs(errors / gold)) * 100
    mape_std = np.std(np.abs(errors / gold)) * 100
    rmse = np.sqrt(np.mean(errors ** 2))
    rmse_std = np.std(errors ** 2) ** 0.5
    p, _ = pearsonr(gold, new)
    return mae, mae_std, mape, mape_std, rmse, rmse_std, p

# =======================
# Bland-Altman 
# =======================
class BlandAltman:
    def __init__(self, gold_std, new_measure, save_path='./bland_altman_plots', averaged=False):
        #  pandas.Series
        if isinstance(gold_std, (list, np.ndarray)):
            gold_std = pd.Series(gold_std)
        if isinstance(new_measure, (list, np.ndarray)):
            new_measure = pd.Series(new_measure)
        self.gold_std = gold_std
        self.new_measure = new_measure

        # 
        diffs = gold_std - new_measure
        self.mean_error = diffs.mean()
        self.std_error = diffs.std()
        self.mean_absolute_error = diffs.abs().mean()
        self.mean_squared_error = (diffs ** 2).mean()
        self.root_mean_squared_error = np.sqrt((diffs ** 2).mean())
        self.correlation = np.corrcoef(self.gold_std, self.new_measure)[0, 1]
        diffs_std = diffs.std()
        corr_std = np.sqrt(2 * (diffs_std ** 2))
        if averaged:
            self.CI95 = [self.mean_error + 1.96 * corr_std, self.mean_error - 1.96 * corr_std]
        else:
            self.CI95 = [self.mean_error + 1.96 * diffs_std, self.mean_error - 1.96 * diffs_std]

        self.save_path = save_path
        os.makedirs(self.save_path, exist_ok=True)

    def print_stats(self, round_amount=5):
        print(f"Mean error = {round(self.mean_error, round_amount)}")
        print(f"Mean absolute error = {round(self.mean_absolute_error, round_amount)}")
        print(f"Mean squared error = {round(self.mean_squared_error, round_amount)}")
        print(f"Root mean squared error = {round(self.root_mean_squared_error, round_amount)}")
        print(f"Standard deviation error = {round(self.std_error, round_amount)}")
        print(f"Correlation = {round(self.correlation, round_amount)}")
        print(f"+95% Confidence Interval = {round(self.CI95[0], round_amount)}")
        print(f"-95% Confidence Interval = {round(self.CI95[1], round_amount)}")

    def rand_jitter(self, arr):
        stdev = .01 * (max(arr) - min(arr))
        return arr + np.random.randn(len(arr)) * stdev

    def scatter_plot(self, x_label='Gold Standard', y_label='New Measure',
                    figure_size=(4, 4), show_legend=True, the_title=' ',
                    file_name='BlandAltman_ScatterPlot.pdf', is_journal=False,
                    measure_lower_lim=40, measure_upper_lim=150):
        if is_journal:
            matplotlib.rcParams['pdf.fonttype'] = 42
            matplotlib.rcParams['ps.fonttype'] = 42

        gold_std_jitter = self.rand_jitter(self.gold_std)
        new_measure_jitter = self.rand_jitter(self.new_measure)
        fig = plt.figure(figsize=figure_size)
        ax = fig.add_axes([0, 0, 1, 1])
        xy = np.vstack([gold_std_jitter, new_measure_jitter])
        z = gaussian_kde(xy)(xy)
        ax.scatter(gold_std_jitter, new_measure_jitter, c=z, s=50)
        x_vals = np.array(ax.get_xlim())
        ax.plot(x_vals, x_vals, '--', color='black', label='Line of Slope = 1')
        ax.set_xlabel(x_label)
        ax.set_ylabel(y_label)
        ax.set_title(the_title)
        ax.grid()
        plt.xlim(measure_lower_lim, measure_upper_lim)
        plt.ylim(measure_lower_lim, measure_upper_lim)
        if show_legend:
            ax.legend()
        plt.savefig(os.path.join(self.save_path, file_name), bbox_inches='tight', dpi=300)
        plt.close()
        print(f"Saved {file_name} to {self.save_path}.")

    def difference_plot(self, x_label='Difference between rPPG HR and ECG HR [bpm]',
                        y_label='Average of rPPG HR and ECG HR [bpm]', averaged=False,
                        figure_size=(4, 4), show_legend=True, the_title='',
                        file_name='BlandAltman_DifferencePlot.pdf', is_journal=False):
        if is_journal:
            matplotlib.rcParams['pdf.fonttype'] = 42
            matplotlib.rcParams['ps.fonttype'] = 42

        diffs = self.gold_std - self.new_measure
        avgs = (self.gold_std + self.new_measure) / 2
        fig = plt.figure(figsize=figure_size)
        ax = fig.add_axes([0, 0, 1, 1])
        xy = np.vstack([avgs, diffs])
        z = gaussian_kde(xy)(xy)
        ax.scatter(avgs, diffs, c=z, label='Observations')
        ax.axhline(self.mean_error, color='black', label='Mean Error')
        ax.axhline(self.CI95[0], color='black', linestyle='--', label='+95% Confidence Interval')
        ax.axhline(self.CI95[1], color='black', linestyle='--', label='-95% Confidence Interval')
        ax.set_ylabel(x_label)
        ax.set_xlabel(y_label)
        ax.set_title(the_title)
        if show_legend:
            ax.legend()
        ax.grid()
        plt.savefig(os.path.join(self.save_path, file_name), bbox_inches='tight', dpi=100)
        plt.close()
        print(f"Saved {file_name} to {self.save_path}.")

    def save_combined_plot(self,
                           diff_x_label='Average of rPPG HR and ECG HR [bpm]',
                           diff_y_label='Difference between rPPG HR and ECG HR [bpm]',
                           scatter_x_label='Gold Standard',
                           scatter_y_label='New Measure',
                           figure_size=(10, 4),
                           diff_title='Difference Plot',
                           scatter_title='Scatter Plot',
                           file_name='BlandAltman_Combined.pdf',
                           measure_lower_lim=40,
                           measure_upper_lim=150,
                           is_journal=False,
                           show_legend=True):
        if is_journal:
            matplotlib.rcParams['pdf.fonttype'] = 42
            matplotlib.rcParams['ps.fonttype'] = 42

        diffs = self.gold_std - self.new_measure
        avgs = (self.gold_std + self.new_measure) / 2
        gold_std_jitter = self.rand_jitter(self.gold_std)
        new_measure_jitter = self.rand_jitter(self.new_measure)
        fig, axes = plt.subplots(1, 2, figsize=figure_size)

        # 
        xy2 = np.vstack([avgs, diffs])
        z2 = gaussian_kde(xy2)(xy2)
        axes[0].scatter(avgs, diffs, c=z2, label='Observations')
        axes[0].axhline(self.mean_error, color='black', label='Mean Error')
        axes[0].axhline(self.CI95[0], color='black', linestyle='--', label='+95% Confidence Interval')
        axes[0].axhline(self.CI95[1], color='black', linestyle='--', label='-95% Confidence Interval')
        axes[0].set_xlabel(diff_x_label)
        axes[0].set_ylabel(diff_y_label)
        axes[0].set_title(diff_title)
        axes[0].grid()
        if show_legend:
            axes[0].legend()

        # 
        xy = np.vstack([gold_std_jitter, new_measure_jitter])
        z = gaussian_kde(xy)(xy)
        axes[1].scatter(gold_std_jitter, new_measure_jitter, c=z, s=50)
        x_vals = np.array(axes[1].get_xlim())
        axes[1].plot(x_vals, x_vals, '--', color='black', label='Line of Slope = 1')
        axes[1].set_xlabel(scatter_x_label)
        axes[1].set_ylabel(scatter_y_label)
        axes[1].set_title(scatter_title)
        axes[1].grid()
        axes[1].set_xlim(measure_lower_lim, measure_upper_lim)
        axes[1].set_ylim(measure_lower_lim, measure_upper_lim)
        if show_legend:
            axes[1].legend()

        plt.tight_layout()
        save_path = os.path.join(self.save_path, file_name)
        plt.savefig(save_path, bbox_inches='tight', dpi=300)
        plt.close()
        print(f" {save_path}")

# =======================
# 
# =======================

# 
gold_std = pd.read_csv('path_to_bvp_hr')['hr']
new_measure = pd.read_csv('path_to_rppg_hr')['hr']
gold_df = pd.read_csv('/data01/km/inference/SUMS_SizeW72_hr.csv')
new_df = pd.read_csv('/data01/km/inference/predict_bvp_hr.csv')

# Bland-Altman 
ba = BlandAltman(gold_std=gold_std, new_measure=new_measure, save_path='./bland_altman_plots', averaged=False)
ba.print_stats()
ba.scatter_plot(
    x_label='BVP -> HR [bpm]', y_label='rPPG -> HR [bpm]',
    figure_size=(4, 4), show_legend=True,
    the_title='Bland-Altman Scatter Plot',
    file_name='BlandAltman_ScatterPlot.pdf', is_journal=False
)
ba.difference_plot(
    x_label='Difference between BVP HR and rPPG HR [bpm]',
    y_label='Average of BVP HR and rPPG HR [bpm]',
    figure_size=(4, 4), show_legend=True,
    the_title='Bland-Altman Difference Plot',
    file_name='BlandAltman_DifferencePlot.pdf', is_journal=False
)
ba.save_combined_plot(
    diff_x_label='Average of BVP HR and rPPG HR [bpm]',
    diff_y_label='Difference between BVP HR and rPPG HR [bpm]',
    scatter_x_label='BVP HR [bpm]',
    scatter_y_label='rPPG HR [bpm]',
    figure_size=(10, 5),
    diff_title='Bland-Altman Analysis',
    scatter_title='Heart Rate Prediction Comparison',
    file_name='BlandAltman_Combined.pdf',
    measure_lower_lim=40,
    measure_upper_lim=150,
    is_journal=False,
    show_legend=True
)

# =======================
# 
# =======================
errors = new_measure - gold_std
mae = np.mean(np.abs(errors))
mae_std = np.std(np.abs(errors))
mape = np.mean(np.abs(errors / gold_std)) * 100
mape_std = np.std(np.abs(errors / gold_std)) * 100
rmse = np.sqrt(np.mean(errors ** 2))
rmse_std = np.std(errors ** 2) ** 0.5
p, _ = pearsonr(gold_std, new_measure)

print(f"all MAE = {mae:.4f}  {mae_std:.4f} bpm")
print(f"all MAPE = {mape:.2f}  {mape_std:.2f} %")
print(f"all RMSE = {rmse:.4f}  {rmse_std:.4f} bpm")
print(f"all Pearson r = {p:.4f}")

# =======================
# 
# =======================
gold_df['file'] = gold_df['file'].str.strip().str.lower()
new_df['file'] = new_df['file'].str.strip().str.lower()

rest_keys = ['v01', 'v03']
recovery_keys = ['v02', 'v04']

gold_rest = extract_group(gold_df, rest_keys)
new_rest = extract_group(new_df, rest_keys)
gold_recovery = extract_group(gold_df, recovery_keys)
new_recovery = extract_group(new_df, recovery_keys)

# rest
mae, mae_std, mape, mape_std, rmse, rmse_std, p = calc_metrics(gold_rest['hr'], new_rest['hr'])
print(f"[Rest] MAE = {mae:.4f}  {mae_std:.4f} bpm")
print(f"[Rest] MAPE = {mape:.2f}  {mape_std:.2f} %")
print(f"[Rest] RMSE = {rmse:.4f}  {rmse_std:.4f} bpm")
print(f"[Rest] Pearson r = {p:.4f}")

# recovery
mae, mae_std, mape, mape_std, rmse, rmse_std, p = calc_metrics(gold_recovery['hr'], new_recovery['hr'])
print(f"[Recovery] MAE = {mae:.4f}  {mae_std:.4f} bpm")
print(f"[Recovery] MAPE = {mape:.2f}  {mape_std:.2f} %")
print(f"[Recovery] RMSE = {rmse:.4f}  {rmse_std:.4f} bpm")
print(f"[Recovery] Pearson r = {p:.4f}")
```
